```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import datetime
%matplotlib inline
```

## Pre procesar los datos

```python
dataset = pd.read_csv("df_depurada_1.csv", encoding = "ISO-8859-1")
dataset.drop(['RADICADO', 'OBJECTID', 'CBML', 'DIRECCION', 'DIRECCION_ENC'], axis=1, inplace=True)
dataset['CANTIDAD'] = dataset.groupby(['FECHA','COMUNA'])['X'].transform('count')
dataset = dataset.sort_values(by=['FECHA', 'COMUNA'], ascending=True)
dataset.reset_index(inplace=True, drop=True) 
dataset.head()
```

```python
entrenamiento_2014 = dataset[dataset['PERIODO'] == 2014]
entrenamiento_2015 = dataset[dataset['PERIODO'] == 2015]
entrenamiento_2016 = dataset[dataset['PERIODO'] == 2016]
entrenamiento = pd.concat([entrenamiento_2014, entrenamiento_2015, entrenamiento_2016], ignore_index=True)
```

```python
clase = pd.get_dummies(data=entrenamiento['CLASE'], drop_first=True)
gravedad = pd.get_dummies(data=entrenamiento['GRAVEDAD'], drop_first=True)
barrio = pd.get_dummies(data=entrenamiento['BARRIO'], drop_first=True)
comuna = pd.get_dummies(data=entrenamiento['COMUNA'], drop_first=True)
diseno = pd.get_dummies(data=entrenamiento['DISENO'], drop_first=True)
dia_nombre = pd.get_dummies(data=entrenamiento['DIA_NOMBRE'], drop_first=True)
tipo_geocod = pd.get_dummies(data=entrenamiento['TIPO_GEOCOD'], drop_first=True)
```

```python
entrenamiento.drop(['CLASE', 'GRAVEDAD', 'BARRIO', 'DISENO', 'DIA_NOMBRE', 'COMUNA', 'TIPO_GEOCOD'], inplace=True, axis=1)
```

```python
entrenamiento_1 = pd.concat([entrenamiento, clase, gravedad, barrio, comuna, diseno, dia_nombre, tipo_geocod], axis=1)
entrenamiento_1.head()
```
