```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import datetime
%matplotlib inline
```

## Pre procesar los datos

```python
dataset = pd.read_csv("df_depurada_1.csv", encoding = "ISO-8859-1")
dataset.drop(['RADICADO', 'OBJECTID', 'CBML', 'DIRECCION', 'DIRECCION_ENC'], axis=1, inplace=True)
dataset['CANTIDAD'] = dataset.groupby(['FECHA','COMUNA'])['X'].transform('count')
dataset = dataset.sort_values(by=['FECHA', 'COMUNA'], ascending=True)
dataset.reset_index(inplace=True, drop=True) 
dataset.head()
```

```python
clase = pd.get_dummies(data=dataset['CLASE'], drop_first=True)
gravedad = pd.get_dummies(data=dataset['GRAVEDAD'], drop_first=True)
barrio = pd.get_dummies(data=dataset['BARRIO'], drop_first=True)
comuna = pd.get_dummies(data=dataset['COMUNA'], drop_first=True)
diseno = pd.get_dummies(data=dataset['DISENO'], drop_first=True)
dia_nombre = pd.get_dummies(data=dataset['DIA_NOMBRE'], drop_first=True)
tipo_geocod = pd.get_dummies(data=dataset['TIPO_GEOCOD'], drop_first=True)
```

```python
dataset.drop(['CLASE', 'GRAVEDAD', 'BARRIO', 'DISENO', 'DIA_NOMBRE', 'COMUNA', 'TIPO_GEOCOD'], inplace=True, axis=1)
entrenamiento_1 = pd.concat([dataset, clase, gravedad, barrio, comuna, diseno, dia_nombre, tipo_geocod], axis=1)
```
